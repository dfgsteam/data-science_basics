````python
import numpy as np
np.random.seed(101)

# Mittelwert und Kovarianzmatrix definieren
mean = np.array([0, 0])
covariance = np.array([[1, 0], [0, 0.2]])

# Anzahl der Samples
num_samples = 2000

# Erstelle den Array mit Samples aus der multivariaten Gaußverteilung
data = np.random.multivariate_normal(mean, covariance, size=num_samples) # Setze den Seed für die Reproduzierbarkeit

# Ausgabe des erstellten Arrays
data
#test = np.load('backup_data_2a.npy')
#print(test)
````

b) [15%] Definieren Sie eine Funktion, die Ihnen eine Rotationsmatrix abhängig von einem Rotationswinkel $\theta$ erstellt

````python
def get_rotation_matrix(theta):

    
    # Calculate cosine and sine of theta
    cos_theta = np.cos(theta)
    sin_theta = np.sin(theta)
    
    # Create the rotation matrix
    rotation_matrix = np.array([[cos_theta, -sin_theta],
                                [sin_theta, cos_theta]])
    
    return rotation_matrix
````

````python
theta = np.radians(40)
r_matrix = get_rotation_matrix(theta)
data_rotated = data @ r_matrix
data_rotated
````
